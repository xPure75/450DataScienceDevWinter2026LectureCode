### The Difference Between `.ipynb` and `.py`
So far, you have probably been using `.py` files to run Python scripts. This is the default file type for Python and is useful for a wide range of tasks, from building web scrapers to creating backend APIs. For many software solutions, `.py` files are exactly what you want.

However, in the world of data science, there is another common file type: `.ipynb`.

`.ipynb` files are primarily used for Jupyter Notebooks. The name originally comes from **IPython Notebook**, which is where the file extension comes from. Today, when we talk about `.ipynb` files, we are generally referring to Jupyter Notebooks.



### Why Not Just Use `.py`?
`.ipynb` files are especially well suited for data analysis and machine learning work. Notebooks combine easy to read Markdown text blocks, like this one, with executable code cells.

Traditional Python projects are often larger and structured across multiple modules and files. Jupyter Notebooks, on the other hand, are better suited for experimentation, exploration, and explaining your thought process alongside your code.
